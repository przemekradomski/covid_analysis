```python
import pandas as pd

import matplotlib.pyplot as plt

# Ścieżka do pliku COVID-19
url = '../data/covid_data.csv'

# Wczytanie danych COVID-19
data = pd.read_csv(url)

# Wyświetlenie pierwszych kilku wierszy danych
print(data.head())

# Wyświetlenie nazw kolumn
print("\nNazwy kolumn:")
print(data.columns.tolist())

# Filtrowanie danych dla konkretnego kraju (np. Polska)
country_data = data[data['Entity'] == 'Poland']

# Wykres liczby zgonów COVID-19 w Polsce w czasie
plt.figure(figsize=(10, 5))
plt.plot(country_data['Day'], country_data['Total confirmed deaths due to COVID-19 per 100,000 people'], label='Zgony COVID-19 (na 100k)')
plt.title('Liczba zgonów COVID-19 w Polsce')
plt.xlabel('Data')
plt.ylabel('Zgony na 100,000 osób')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
