```python
import pandas as pd

import matplotlib.pyplot as plt

# Wczytanie danych COVID-19
url = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/jhu/covid19.csv'
data = pd.read_csv(url)

# Wyświetlenie pierwszych kilku wierszy danych
print(data.head())

# Filtrowanie danych dla konkretnego kraju (np. Polska)
country_data = data[data['country'] == 'Poland']

# Wykres liczby przypadków w czasie
plt.figure(figsize=(10, 5))
plt.plot(country_data['date'], country_data['cases'], label='Liczba przypadków')
plt.title('Liczba przypadków COVID-19 w Polsce')
plt.xlabel('Data')
plt.ylabel('Liczba przypadków')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
